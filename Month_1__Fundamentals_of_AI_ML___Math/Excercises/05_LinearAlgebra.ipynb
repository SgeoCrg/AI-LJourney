```python
matrix = np.arange(1, 31).reshape(6, 5)
print(matrix)
subMatrix = matrix[2:4, 0:2] # Remember start => end + 1, [row, col]
print(subMatrix)
sub2 = matrix[[0,1,2,3],[1,2,3,4]] # pares (0,1),(1,2),(2,3),(3,4)
print(sub2)
sub3 = matrix[[0,4,5], 3:] # row 0,4 & 5 position 3 till the end (5)
print(sub3)
```
