```python
import numpy as np
```

```python
a = np.array([1,2,3])
print(a)
```

```python
b = np.array([[9.0, 8.0, 7.0], [6.0, 5.0, 4.0]])
print(b)
```

```python
#Get dimension
print(a.ndim)
b.ndim
```

```python
#Get Shape
print(a.shape)
b.shape
```

```python
#Get type
print(a.dtype)
b.dtype
```

```python
a2 = np.array([1,2,3], dtype='int16')
a2.dtype
```

```python
#Get size
print(a.itemsize)
a2.itemsize
```

```python
#Get total size
print(a.size * a.itemsize)
b.size * b.itemsize
```
