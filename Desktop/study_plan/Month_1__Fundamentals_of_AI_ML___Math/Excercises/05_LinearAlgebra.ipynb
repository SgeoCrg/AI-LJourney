```python
stats = np.array([[1, 2, 3], [4, 5, 6]])
```

```python
print(np.max(stats))
np.max(stats, axis = 1)
```

```python
print(np.sum(stats))
np.sum(stats, axis = 1)
```

```python
# Vertically stacking vectors
v1 = np.array([1, 2, 3, 4])
v2 = np.array([5, 6, 7, 8])

np.vstack([v1, v2, v1, v2])
```

```python
h1 = np.ones((2, 4))
h2 = np.zeros((2, 2))
np.hstack([h1, h2])
```
